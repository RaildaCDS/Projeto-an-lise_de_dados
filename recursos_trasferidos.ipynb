### Renomeando os as Colunas

```python
recursos_transferidos.rename(columns={'Ação Orçamentária': 'Acao_Orcamentaria'}, inplace = True)
```

```python
recursos_transferidos.rename(columns={'Valor Transferido': 'Valor_Transferido'}, inplace = True)
```

```python
recursos_transferidos.columns
```

```python
#Visualizando o numero de linhas e colunas.
recursos_transferidos.shape
```

```python
recursos_transferidos.dtypes
```

```python
#Formatação da coluna Varlor_Transferido para transformar no tipo float.
recursos_transferidos['Valor_Transferido'] = recursos_transferidos['Valor_Transferido'].apply(lambda x: str(x).replace("R$",""))
recursos_transferidos['Valor_Transferido'] = recursos_transferidos['Valor_Transferido'].apply(lambda x: str(x).replace(" ",""))
recursos_transferidos['Valor_Transferido'] = recursos_transferidos['Valor_Transferido'].apply(lambda x: str(x).replace(".",""))
recursos_transferidos['Valor_Transferido'] = recursos_transferidos['Valor_Transferido'].apply(lambda x: str(x).replace(",","."))
recursos_transferidos['Valor_Transferido'] = recursos_transferidos['Valor_Transferido'].apply(lambda x: str(x).replace("R$",""))
```

```python
#Convertendo os valores da coluna Valor_transferido para Float64.
recursos_transferidos['Valor_Transferido'] = recursos_transferidos['Valor_Transferido'].astype('float64')
```

```python
#Formatando a coluna Valor_transferido como float para exibir dois dígitos após o ponto.
pd.options.display.float_format = '{:.2f}'.format
```
